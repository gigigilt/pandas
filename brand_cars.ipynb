```python
#1-1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
brand_df=pd.read_csv("/aiffel/aiffel/Untitled Folder/brand.csv")
brand_df['title']=brand_df['title'].str.upper()
brand_df
```

```python
cars_df=pd.read_csv("/aiffel/aiffel/Untitled Folder/cars.csv")
cars_df
cars_df['title']=cars_df['title'].str.split().str[0]
cars_df
```

```python
new_cars_df=cars_df.merge(brand_df, on="title", how='left')
new_cars_df
```

```python
#1-2
new_cars_df.info()
```

```python
new_cars_df['Engine']=new_cars_df['Engine'].str.split('L').str[0]
new_cars_df['Emission Class']=new_cars_df['Emission Class'].str.split().str[-1]
encoded_cars_df=new_cars_df
encoded_cars_df
```

```python
#1-3
encoded_cars_df.isna().sum()
```

```python
encoded_cars_df['Service history']=encoded_cars_df['Service history'].fillna('Unknown')
newenc_cars_df=encoded_cars_df.copy()
newenc_cars_df[newenc_cars_df['Service history']!='']
```

```python
encoded_cars_df=encoded_cars_df.drop(['Previous Owners','country', 'Doors', 'Seats'], axis=1)
encoded_cars_df[~(encoded_cars_df['Gearbox']=='Manual')]
print(encoded_cars_df['title'].nunique())
print(encoded_cars_df['Fuel type'].nunique())
print(encoded_cars_df['Body type'].nunique())
```

```python
scaler_cars_df=pd.get_dummies(encoded_cars_df, columns=['title','Fuel type', 'Body type','Gearbox', 'Service history'], drop_first=True)
print(scaler_cars_df)
```

```python
#1-4
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
ss=StandardScaler()
mm=MinMaxScaler()
rb=RobustScaler()
ss_s=ss.fit_transform(scaler_cars_df)
ss_s
```

```python
mm_s=mm.fit_transform(scaler_cars_df)
mm_s
```

```python
rb_s=rb.fit_transform(scaler_cars_df)
rb_s
```

```python
#1-5
mm_s_df=pd.DataFrame(mm_s, columns=scaler_cars_df.columns)
print(mm_s_df.info())
print(mm_s_df.isna().sum())
mm_s_df=mm_s_df.dropna()
print()
mm_s_df.isna().sum()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.7)
pca_result=pca.fit_transform(mm_s_df)
pca_result
```

```python
pca.explained_variance_ratio_
```

```python
print(pca_result)
print('축소된 차원의 수',':', pca.n_components_)
print('설명된 분산 퍼센트', ':', pca.explained_variance_ratio_)
print('누적된 설명 분산 퍼센트',':' , sum(pca.explained_variance_ratio_*100))
```
